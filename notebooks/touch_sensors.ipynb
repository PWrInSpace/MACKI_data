```python
import pandas as pd
import glob
import os
```

```python
macus_data = glob.glob("../**/MACUS*.TXT", recursive=True)
```

```python
sensors_header = ["sensor_" + str(i) for i in range(0, 32)]
data_header = ["uptime"] + sensors_header
data_header
```

```python
df_list = []
for macus_file in macus_data:
    df = pd.read_csv(macus_file, sep=';', names=data_header, index_col=False)
    df["file"] = os.path.basename(macus_file)
    df_list.append(df)

df = pd.concat(df_list)
df
```

```python
df = df[(df != 0).all(axis=1)]
df
```

```python
df_sensors = df[sensors_header]

outliers = df_sensors[(df_sensors < 8380000) | (df_sensors > 8400000)].dropna(how='all')
outliers
```
